```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
plt.plot(x, y, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.title('plot')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
plt.show()
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axesi = fig.add_axes([0.2, 0.6, 0.25, 0.25])

axes.plot(x, y, 'r--')
axes.set_xlabel('x original')
axes.set_ylabel('y original')
axes.set_title('title original')

axesi.plot(y, x, '--g')
axesi.set_xlabel('x inverted')
axesi.set_ylabel('y inverted')
axesi.set_title('title inverted')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
b = 4
for ax in axes:
    ax.plot(x, y, "#A501FC")
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    b = b+1

fig    
plt.tight_layout()
```
